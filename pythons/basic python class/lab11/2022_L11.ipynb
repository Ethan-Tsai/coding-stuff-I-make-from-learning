## Matplotlib 圖表

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

簡單來說，簡單他就是個好用的圖表 library，也是很多圖表套件的後端，是資料分析必備的套件之一

### 基礎圖表

先來 import 吧，他跟 numpy 一樣也有通用的格式

````python
import numpy as np # 製作資料
import matplotlib.pyplot as plt
````

接著可以使用幾種方式來做圖表
先來一個最簡單的
```python
plt.plot(data)
plt.plot([1, 2, 5, 6, 10])
```
data 可以是 list 也可以是 numpy.array，這邊的會用最簡單的「折線圖」顯示資料

注意：如果你是在 pc 使用，要額外呼叫 plt.show() 才會真的顯示出來  
而在 notebook 裡面則可以省略

````python
plt.plot([1, 2, 4, 6, 10])
````

你也可以輸入 x 和 y
```python
plt.plot(x_data, y_data)
plt.plot([1, 2, 3, 9, 10], [2, 3, 6, 81, 100])
```

````python
plt.plot([1, 2, 3, 8, 9, 10], [2, 3, 6, 64, 81, 100])
````

````python
# 延續上次作業
x = np.linspace(0, np.pi*4, 51)
y = np.sin(x)
plt.plot(x, y)
````

你可以用 color= 或 c= 傳顏色進去

````python
data = [1, 2, 4, 6, 10]
````

````python
plt.plot(data, color='g')
# 或 c='g' 也可以
````

````python
# 或色碼
plt.plot(data, color='#9900ff')
````

````python
# 疊兩條線
plt.plot(x, y)
plt.plot(x, np.cos(x), c='g')
````

#### 一些其他圖表

另外還有 scatter 散點圖、histogram 長條圖 這兩種都是常用的圖表類型  
分別呼叫 `plt.scatter` 和 `plt.hist` 就可以繪製

```python
plt.scatter(x, y)
```

```python
plt.hist(data)
```

````python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
````

````python
data['a']
````

````python
data['b']
````

````python
plt.scatter(data['a'], data['b'])
````

````python
# 進階散點圖，使用 a b 作為點的位置、c 作為顏色、d 作為 size 傳入
plt.scatter('a', 'b', c='c', s='d', data=data)
````

````python
# 顯示 b 的數值分佈
plt.hist(data['b'])
````

````python
size = [10, 3, 6]
plt.pie(size, labels=['a', 'b', 'c'], autopct='%.02f%%')
plt.legend()
````

### 圖表樣式

plot 可以有一個參數叫 fmt  
可以傳入最多三個字元的設定  
他可以快速的指定 marker, line 和 color 的樣式

```python
fmt = '[marker][line][color]'
```

#### marker 樣式

| 字串 | 樣式 |
| :---: | ---- |
| . | point marker |
| o | circle marker |
| ^ | triangle_up marker |
| + | plus marker |

#### line 樣式

| 字串 | 樣式 |
| :---: | ---- |
| - | solid line style |
| -- | dashed line style |
| -. | dash-dot line style |
| : | dotted line style |

#### color

| 字串 | 樣式 |
| :---: | ---- |
| - | solid line style |
| -- | dashed line style |
| -. | dash-dot line style |
| : | dotted line style |

````python
plt.plot(x, y, 'b--')
````

````python
plt.plot(x, y, 'go-.')
````

````python
# scatter 只能用 keyword 傳進去，因為他沒有 line
plt.scatter(data['a'], data['b'], marker='^')
````

````python
plt.scatter('a', 'b', c='c', s='d', data=data, marker='+')
````

除此你也可以設定圖以外的東西

![](https://miro.medium.com/max/2400/0*1s77sR1nFozVzX8W.png)

plt 把圖分成這幾個區塊  
你可以對外圍的這些樣式做修改

````python
plt.plot(x, y)
plt.plot(x, np.cos(x), c='g')
plt.legend(['sin', 'cos']) # 顯示 圖例
plt.title('sin cos Plot', c='g') # 有些也能傳參數進去
````

````python
plt.plot(x, y)
plt.plot(x, np.cos(x), c='g')
plt.ylabel('hi there')
plt.ylim(-2, 2)
plt.legend(['sin', 'cos']) # 顯示 圖例
````

````python
with plt.style.context('dark_background'):
    plt.hist(data['b'])
    plt.xlim(-20, 100)
    plt.title('hihi', size=30, color='b')
````

````python
plt.scatter('a', 'b', c='c', s='d', data=data, marker='+')
plt.xlim(10, 60) # 修改 x 軸範圍
plt.ylim(-20, 60) # 修改 y 軸範圍
````

````python
plt.scatter(data['a'], data['b'], marker='^')
plt.xlabel('x label')
plt.ylabel('y label')
````

````python
# 印出所有樣式
print(plt.style.available)
````

````python
plt.style.use('fivethirtyeight') # 全部都會影響到
plt.scatter('a', 'b', c='c', s='d', data=data, marker='+')
````

````python
# 暫時
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
````
