```python
class GamePlayer():
    # create a new game player
    def __init__(self, name):
        self.name = name
        self._max_health = 20
        self._max_mana = 100
        self.health = 20
        self.mana = 100
        self._strength = 100
        self.level = 1
        self.alive = True
        print(f'{self.name} join the world')

    # delete a player
    def __del__(self):
        print(f'{self.name} leave the world')

    # level up method
    def level_up(self, lvl):
        self._max_health += 5 * lvl
        self._max_mana += 10 * lvl
        self.health = self._max_health
        self.mana = self._max_mana
        self._strength += 10 * lvl
        self.level += lvl
        print(f'{self.name} level up to {self.level}')

    # heal method
    def heal(self):
        # check if the player is dead
        if self.alive == False:
            self.alive = True
            print(f'{self.name} revive')
            
        # check to prevent over the max health
        if self.health + int(float(self._max_health) / 10) > self._max_health:
            self.health = self._max_health
        else:
            self.health += int(float(self._max_health) / 10)

        # check to prevent over the max mana
        if self.mana + int(float(self._max_mana) / 10) > self._max_mana:
            self.mana = self._max_mana
        else:
            self.mana += int(float(self._max_mana) / 10)

    # attack method
    def melee_attack(self, other, dmg: float):
        print(
            f'{self.name} impact to {other.name} deal {int(float(self._strength) / 100 * dmg)} melee dmg')
        other.health -= int(float(self._strength) / 100 * dmg)
        # check if the health is below 0
        if other.health < 0:
            print(f'{other.name} is dead')
            other.health = 0
            other.alive = False

    def mage_attack(self, other, dmg: float):
        # check is this attack valid
        if self.mana > dmg / 10:
            print(
                f'{self.name} impact to {other.name} deal {int(float(self.mana) / 100 * dmg)} mage dmg')
            other.health -= int(float(self.mana) / 100 * dmg)
            self.mana -= int(dmg / 10)

            # check if the health is below 0
            if other.health < 0:
                print(f'{other.name} is dead')
                other.health = 0
                other.alive = False

    # show progress
    def show_stats(self):
        print(f'\n{self.name} :')
        print(f'- health : {self.health} / {self._max_health}')
        print(f'- mana : {self.mana} / {self._max_mana}\n')


# main code
Steve = GamePlayer('Steve')
Alex = GamePlayer('Alex')

Alex.show_stats()
Steve.show_stats()

Alex.level_up(10)
Steve.level_up(100)

Alex.show_stats()
Steve.show_stats()

Alex.mage_attack(Steve, 100)

Alex.show_stats()
Steve.show_stats()

Alex.melee_attack(Steve, 1000)

Alex.show_stats()
Steve.show_stats()

Steve.heal()

Steve.show_stats()

del Steve, Alex
```
