### Aggregate, filter, transform and apply

```python
# Create a dataframe

df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': range(6),
    'data2': np.random.randint(0, 10, 6)
})

print(df)
```

- We want to perform a set of aggregate functions
- 'aggregate()' provide us such flexibility

```python
df.groupby('key').aggregate(['min', np.median, max])
```

**What if we want to perform specific function to a particular column**

- Creating a dictionary and mapping the column names and function names

```python
df.groupby('key').aggregate({
    'data1': min,
    'data2': max
})
```

### Filter

```python
df
```

```python
df.groupby('key').std()
```

- Creating the condition which keeps the values in a row by comparing the standard deviation in data2

```python
def filter_func(x):
    return x['data2'].std() > 4
```

```python
df.groupby('key').filter(filter_func)
```

```python
df.groupby('key').filter(lambda x: x['data2'].std() > 4)
```

### Tranformation

- Aggregation reduced the version of data
- Transformation return some transformed version of the full data to recombine

- For example to center the data by substracting the group-wide mean

```python
display(df, df.groupby('key').mean())
```

```python
# Subtracting the row value of a column by group mean
df.groupby('key').transform(lambda x: x - x.mean())
```

### Apply

```python
df
```

```python
# Create the function of arbitrary function to normalize the column one

def normal_by_data(x):
    x['data1'] /= x['data2'].sum()
    return x
```

```python
df.groupby('key').apply(normal_by_data)
```

- Creating the custom keys for groupby

```python
L = [0, 1, 0, 1, 2, 0]

display(df, df.groupby(L).sum())
```

- Create a dictionary that map index_values to a group keys

```python
df2 = df.set_index('key')
print(df2)
```

```python
mapping = {'A': 'vowel',
           'B': 'consonant',
           'C': 'consonant'
}

df2.groupby(mapping).sum()
```
