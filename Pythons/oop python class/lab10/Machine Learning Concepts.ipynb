# Categories of Machine Learning

**`Supervised learning`**
- Involves somehow modeling the relationship between measured features of data and some label 
associated with the data
- Once this model is determined
- Model can be used to apply labels to new, unknown data. 

**Supervised learning further divided in to `classification` task and `regression` task**
- In classification, the labels are `discrete categories`
- In regression, the labels are `continuous quantities`

**`Unsupervised learning**` 
- Involves modeling the features of a dataset without reference to any label 
- It is often described as "letting the dataset speak for itself." 

**Unsupervised learning further divided in to `clustering` task and `dimensionality reduction` task**

- Clustering algorithms identify distinct groups of data
- Dimensionality reduction algorithms search for more concise representations of the data

**Semi-supervised learning methods**
- Falls somewhere between supervised learning and unsupervised learning. 
- Semi-supervised learning methods are often useful when only incomplete labels are available.

## Scikit-Learn's Estimator API

### Basics of the API

**Steps followed to use Scikit-Learn Estimator API**
- Choose a class of model 
- Choosee a model hyperparameters
- Arrange data into a `features matrix` and `target vector`
- Fit the model to your data by calling `fit()` method of the model instance

    - For `supervised learning`, we predict labels for unknow data using the `predict()` method

### Supervised learning example: Simple linear regression

- Consider a simple linear regression by fitting a line to (x,y) data
- We will use Iris simple data for our regression example

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50) # 2 is slop and -1 is intercept
plt.scatter(x, y)
print(x.shape)
print(x.ndim)
print(y.shape)
print(y.ndim)
```

- Now let's follow the steps outlied earlier

1. Choose a class model
    - we can import a linear regression class

```python
from sklearn.linear_model import LinearRegression
```

2. Choose a model hyperparameter
    - A class of model is not same as an instance of a model
    <br></br>
    - Before that we need to ask following questions
    1. Would we like to fit for the offset (i.e. intercept)?
    2. Would we like the model to be normalized?
    3. Would we like to preprocess our features to add model flexibility?
    4. What degree of regularization would we like to use in our model?
    5. How many model components would we like to use?
    <br></br>
    - Above choices are often represented by `hyperparameters`
    - These parameters must be set before the model is fit to data
    <br></br>
    - We can choose hyperparameters by passing values at model instantiation

```python
# Instantiation of LinearRegression class and specifying the hyperparameter
# The only action here is storing the value of these hyperparameter
model = LinearRegression(fit_intercept=True)
model
```

3. Arrange data into `features matrix` and `target vector`
    - Requirement of two-dimensional features matrix and one-dimensional target array
    - Our target variable is already in shape ( a length n-samples array)
    - But we need to transform our x variable in to a matrix [n_samples, n_features]
    - Reshaping of one-dimensional array

```python
X = x[:, np.newaxis]
print(X.shape)
print(X.ndim)
#print(X)
```

4. Fit the model to your data
    - applying our model to data i.e. using fit() method of the model

```python
model.fit(X, y)
```

- The results are stored in model -specific attributes
- All model parameters that were learned during the fit() process have trailing underscore

```python
model.coef_
```

```python
model.intercept_
```

- The resultant coef and intercept are very close to our definition of slop 2 and intercept -1

5. `Predict labels` for `unknown data`
    - Here new data will be a grid of x values
    - We will ask what y values the model predicts

### Model validation via cross-validation

- One disadvantage of holdout ser for model validation is that we have lost a portion of our data to the model training
- It will be problematic if the initial data is small

- One way to address this is to use `cross-validation`
- In cross-validation each subset of data is used as a training set and a validation set

<img src='figure_two_fold_cross_validation.png' align='left'>

- Figure two fold cross validation

- Alternatively we can use each half of the data as a holdout set.
- We can use split data

```python
from sklearn.metrics import accuracy_score
```

```python
y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

- We got two accuracy scores which can be combine to get the mean for a better measurement of 
global model performance
- This form of cross-validation is a two-fold cross-validation
- Where in we divided the data into two set and for each turn each set is used as a validation set

### error: Validation curves in Scikit-learn

- Calculating validation curve using cross-validation
- We will be using ploynomial regression which is a generalized linear model in which degree of polynomial is a tunable parameter.
- For example 1-degree of polynomial fits a straight line to the data
    - $y = \beta_{0}+ \beta_{1}x$
- 3-degree polynomial model parameter $\beta_{0},\; \beta_{1},\; \beta_{2},\; \beta_{3}$
    - $y = \beta_{0}+ \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3}$

- To generalize the model we will be using number of polynomial features
- We can combine simple linear regression with polynomial preprocessor
- We will use pipeline to string these operations together

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                        LinearRegression(**kwargs))
```

- Creating data to fit the model

```python
import numpy as np

def make_data(N, err=1.0, rseed=1):
    #randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2 # Random values in a given shape i.e. N values of 1-dimension
    y = 10 - 1 / (X.ravel() + 0.1) # X.ravel() Return a flattened array.
    if err > 0:
        y += err * rng.randn(N)
        return X, y
    
X, y = make_data(40)
```

**Visualize the data with polynomial fits of several degree**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() #plot formatting
```

```python
X_test = np.linspace(-0.1, 1.1, 500)[:,None]

plt.scatter(X.ravel(), y, color='black')
axis = plt.axis()
for degree in [1, 3, 5]:
    y_test = PolynomialRegression(degree).fit(X,y).predict(X_test)
    plt.plot(X_test.ravel(), y_test, label='degree={0}'.format(degree))
    
plt.xlim(-0.1, 1.0)
plt.ylim(-2,12)
plt.legend(loc='best')
```

- Here the model complexity is controlled by polynomial degree.

- Now the question is what degree of polynomial is provideing trade-off between bias (underfitting) and variance (overfitting)

- We can answer above question by visualizing the validation curve for this particular data and model
- Using `validation_curve` convenience routine provided by scikit-learn
- It calculates the training and testing(validation) score across the range
- we need to input model, data, parameter name and a range to explore

```python
#from sklearn.learning_curve import validation_curve
from sklearn.model_selection import validation_curve    
#ModuleNotFoundError: No module named 'sklearn.learning_curve'
```

```python
from sklearn.model_selection import validation_curve
degree = np.arange(0, 21)
train_score, val_score = validation_curve(estimator=PolynomialRegression(), X=X, y=y,
                                          param_name='polynomialfeatures__degree', param_range=degree, cv=7)

plt.plot(degree, np.median(train_score, 1), color='blue', label='training score')
plt.plot(degree, np.median(val_score, 1), color='red', label='validation score')
plt.legend(loc='best')
plt.ylim(0, 1)
plt.xlabel('degree')
plt.ylabel('score');
```

- The peak here is around a degree of 6, but even a degree -20 model is not overfitting the data.
- The validation and training scores remains very close.

- The validation curve is having two important inputs
    - model complexity (number of polynomial) and number of training points
    

- A plot of training and validation score with respect to the size of the training set is known as a `learning curve`

**- General behavior of a learning curve**
- Given a general model complexity for a smaller dataset the model will have a overfit where in training score 
will be relative higher, while the validation score will be relatively low.
- Given a general model complexity for a larger dataset the model will have a underfit where in training score 
will be relative lower, but the validation score will increase.
- A model will never, except by chance, give a better score to the validation set than the training set.

### Learning curves in Scikit-Learn

- Let's compute learning curve using sk-learn library 
- Where in we will be using second-order and ninth-order polynomial model
- Here in we have used our origial dataset of 40 datapoints

```python
from sklearn.model_selection import learning_curve

fig, ax = plt.subplots(1, 2, figsize = (16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for i, degree in enumerate([2, 9]):
    N, train_lc, val_lc = learning_curve(PolynomialRegression(degree),
                                        X, y, cv=7, 
                                        train_sizes=np.linspace(0.3, 1, 25))
    
    ax[i].plot(N, np.mean(train_lc, 1), color='blue', label='training score')
    ax[i].plot(N, np.mean(val_lc, 1), color='red', label='validation score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1]]), N[0], N[-1],
                color='gray', linestyle='dashed')
    
    ax[i].set_ylim(0, 1)
    ax[i].set_xlim(N[0], N[-1])
    ax[i].set_xlabel('training size', size=20)
    ax[i].set_ylabel('score', size=20)
    ax[i].set_title('degree={0}'.format(degree), size=20)
    ax[i].legend(loc='best')
```

- This visualization of learning curve gives us the depiction of how our model is responding to increasing the
training dataset.
- Adding more training dataset will not significantly improve the fit.

- Other way to increase the converged score is to use a more complecated model i.e. increasing degree of polynomial
but this will be at the expense of higher model variance
- If we add more data points, the learning curve for the more complicated model eventually converge.

- `Learning curve` help us to make the decision on how to move forward in improving our analysis.

## Derived Features

- Means the features that are mathematically derived from some input features
- Like converting a linear regression into polynomial regression from input data

- Example:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 2, 1, 3, 7])
plt.scatter(x, y);
```

- above data can't be well describe by a straight line

- Trying to fit straight line to get the optimal result

```python
from sklearn.linear_model import LinearRegression


X = x[:, np.newaxis]
model = LinearRegression().fit(X, y)
yfit = model.predict(X)
plt.scatter(x,y)
plt.plot(x, yfit)
```

- A poor fit of a straight line

- We need a more complex model to describe the relationship between x and y.
- To drive more flexibility in the model we can add extra polynomial features means adding extra features 

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, include_bias=False)
X2 = poly.fit_transform(X)
print(X)
print(X2)
```

- Now lets compute the linear regression on this data

```python
model = LinearRegression().fit(X2, y)
yfit = model.predict(X2)
plt.scatter(x, y)
plt.plot(x, yfit)
```

- This is a linear fit to polynomial features

- This is the idea of improving the model nit by changing the model but by transforming the inputs.

## Imputation of Missing Data

- How to handle missing data using machine learning approach
- Using imputation of missing values, range from mean, median or most frequent values

```python
from numpy import nan

X = np.array([[nan, 0, 3],
             [3, 7, 9],
             [3, 5, 2],
             [4, nan, 6],
              [8, 8, 1]]
            )

y = np.array([14, 16, -1, 8, -5])
```

```python
#from sklearn.preprocessing import Imputer
# import error
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=nan, strategy='mean')
X2 = imp.fit_transform(X)
X2
```

```python
#most_frequent
imp = SimpleImputer(missing_values=nan, strategy='most_frequent')
X2 = imp.fit_transform(X)
X2
```

```python
#constant by default filled with zero
imp = SimpleImputer(missing_values=nan, strategy='constant')
X2 = imp.fit_transform(X)
X2
```

```python
#constant with fill value it needs to fill desired value
imp = SimpleImputer(missing_values=nan, strategy='constant', fill_value=6)
X2 = imp.fit_transform(X)
X2
```

- The imputed data can be fed into LinearRegression estimator

```python
model = LinearRegression().fit(X2, y)
model.predict(X2)
```

## Feature Pipelines

- We have done multiple steps like
    - Imputing missing values using mean
    - Transforming features to quadratic
    - Fit a linear regression

- How about string together those multiple steps?
- To streamline this type of processing pipeline Scikit-Learn provide a pipeline object
- Using `make_pipeline`


```python
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(SimpleImputer(strategy='mean'),
                     PolynomialFeatures(degree=2),
                     LinearRegression())
```

- This pipeline looks and acts like a standard Scikit-Learn object 

```python
model.fit(X, y) # X from missing values from above
print(y)
print(model.predict(X))
```
