### Seaborn: Example of histogram and density plot

```python
comp1 = np.random.normal(0, 1, size=200)
# The value mean = 0 std = 1 and size is 200 which is normal distributed
```

```python
comp2 = np.random.normal(10, 2, size=200)
comp3 = np.random.normal(10, 2, size=10)
```

```python
comp3
```

```python
sns.displot(values, bins=100, color='g', kde=True)
```

### Scatter plot or point plot

```python
macro = pd.read_csv('macrodata.csv')
macro
```

- CPI: consumer price index
- m1: nominal money stock
- tbilrate: 3-month treasure bill
- umemp: unemployment rate

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data
```

### Calculate percentge change

- An alternative to using percentge change is "log diferent"
- In this method,the loganthm of one quantity is subtracted from the loganthm of another
- The advantage of measuring change in this way is that the calculations are symmetrical going forward and backward

```python
trans_data = np.log(data).diff().dropna()
trans_data
```

### Using scatter plot

- Scatter plot is a useful way to examine the relatinship between two one-dimensional series
- We use seaborn regplot method which make scatter plot and fits a linear regression line

```python
# Figsize adjust as per requirment
fig,ax = plt.subplots(figsize = (8,6))
sns.regplot(x='m1',y='unemp',data=trans_data,ax=ax)
```

## Pairplot or scatter plot matrix

- In exploratory data analysis it is helpful to look at all scatter plots among a group of variables
- This is known as pair plot or scatter plot matrix
- We use seaborn's pairplot function which places histogram or density estimates of each variable along the diagonal

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.7})
```

```python
sns.pairplot(trans_data, diag_kind='hist', plot_kws={'alpha': 0.7})
```

### Facet Grid and Categorical Data

- What if we have additional grouping dimensions
- To visualize many categorical variables is to use facet grid
- Seaborn's factorplot provides such functions to plot faceted plot

```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips['tip_pct'] < 1])
```

```python
sns.catplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips['tip_pct'] < 1])
```

### Box plot

- What if we want to visualize median, quantile, and outliers
- Box plot can be effective way to visualize such type of data

```python
sns.catplot(x='tip_pct', y='day', kind='box', data=tips[tips['tip_pct'] < 0.5])
```
